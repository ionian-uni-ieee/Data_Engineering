<h2>Basic API Usage</h2>

```python
import requests
api_url = "https://gorest.co.in/public/v2/users"

response = requests.get(api_url, headers={"API_TOKEN": "1234"})
display(response.json())
```

```python
api_url = "https://gorest.co.in/public/v2/users/3858"
response = requests.get(api_url)
response.json()
```

```python
api_url = "https://gorest.co.in/public/v2/users?status=active"
response = requests.get(api_url)
print(response.json())
```

```python
api_url = "https://gorest.co.in/public/v2/users?status=active&gender=male"
response = requests.get(api_url)
print(response.json())
```
